**Expression:**

$\theta = \underset{\theta}{argmax}[Pr(x_1...x_n|\theta)]$

```python
import matplotlib.pyplot as plt
import numpy as np
import math
from scipy import stats
```

```python
# Prepare data
original_mu = 5
original_sig = 8
r = original_mu + original_sig * np.random.randn(100,1)
```

```python
def mleNorm(x):
    I = x.size;
    mu = sum(x) / I;
    var = sum(pow(x - mu, 2)) / I;
    return mu, var
```

```python
estimated_mu, estimated_var = mleNorm(r)
estimated_sig = np.sqrt(estimated_var)

# Error value
muError = abs(original_mu - estimated_mu)
sigError = abs(original_sig - estimated_sig)
print('Error value: ', muError,sigError)
```

```python
# Normalize distribution
x = np.linspace(-20,30,100)
original = stats.norm.pdf(x, original_mu, original_sig)
estimated = stats.norm.pdf(x, estimated_mu, estimated_sig)
```

```python
plt.plot(x, original, label = 'Original')
plt.plot(x, estimated, label = 'Estimated')
plt.xlabel('X')
plt.ylabel('P(x)')
plt.legend()
plt.show()
```
